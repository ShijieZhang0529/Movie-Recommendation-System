```python
import pandas as pd
from sklearn import preprocessing
```

```python
df = pd.read_csv('../data/ml-latest-small/movies.csv')
df.head()
```

```python
df.genres = df.genres.str.split('|')
```

```python
df.head()
```

```python
genre_type = list(set([j for i in df['genres'].tolist() for j in i]))
for j in genre_type:
    df[j] = 0
for i in range(df.shape[0]):
    for j in genre_type:
        if (j in df["genres"].iloc[i]):
            df.loc[i,j] = 1
```

```python
df.drop('genres',axis=1,inplace=True)
df.head()
```

```python
df.to_csv('../output/genres.csv')
```

```python
ratings = pd.read_csv('../data/ml-latest-small/ratings.csv')
ratings.drop('timestamp',axis=1,inplace=True)
movieid = ratings.loc[:,"movieId"]
le = preprocessing.LabelEncoder()
le.fit(movieid)
movieid_reindex = le.transform(movieid)
ratings.loc[:,"movieId"] = movieid_reindex + 1
ratings.head()
```

```python
result = pd.merge(ratings, df, how='left',on='movieId')
result.head()
```

```python
result.to_csv('../output/movie_genre.csv')
```
